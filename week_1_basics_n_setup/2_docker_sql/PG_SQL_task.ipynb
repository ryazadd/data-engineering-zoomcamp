```python
df.lpep_pickup_datetime = pd.to_datetime(df.lpep_pickup_datetime)
df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime)
```

```python
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

```python
print(pd.io.sql.get_schema(df, name='green_taxi_data', con=engine))
```

```python
result_set = engine.execute("""
SELECT COUNT(*) FROM zoomcamp.green_taxi_data 
WHERE DATE(lpep_pickup_datetime) ='2019-01-15' 
and DATE(lpep_dropoff_datetime)='2019-01-15';
""")  
for r in result_set:  
    print(r)
```

```python
result_set = engine.execute("""
SELECT date(lpep_pickup_datetime)
FROM zoomcamp.green_taxi_data 
WHERE trip_distance = (SELECT MAX(trip_distance) FROM zoomcamp.green_taxi_data);

""")  
for r in result_set:  
    print(r)
```

```python
result_set = engine.execute("""
SELECT passenger_count, count(*)
FROM zoomcamp.green_taxi_data 
WHERE passenger_count in (2,3) and DATE (lpep_pickup_datetime)='2019-01-01'
GROUP BY passenger_count;
""")  
for r in result_set:  
    print(r)
```

```python
result_set = engine.execute("""
SELECT gx."DOLocationID", tz2."Zone" , tip_amount
FROM zoomcamp.green_taxi_data gx
LEFT JOIN zoomcamp.taxi_zone_lookup tz ON gx."PULocationID"=tz."LocationID"
LEFT JOIN zoomcamp.taxi_zone_lookup tz2 ON gx."DOLocationID"=tz2."LocationID"
WHERE tz."Zone" ='Astoria'
ORDER BY 3 DESC
LIMIT 1
;
""")  
for r in result_set:  
    print(r)
```

```python
result_set = engine.execute("""
SELECT "DOLocationID"
FROM zoomcamp.green_taxi_data 
WHERE tip_amount = ( SELECT MAX(tip_amount)
FROM zoomcamp.green_taxi_data ) ;
""")  
for r in result_set:  
    print(r)
```
